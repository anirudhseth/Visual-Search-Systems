```python
import numpy as np
from utils import *
```

```python
sift_features_with_id=np.load('server_dict_no_kp.npy',allow_pickle=True).item()
sift_features_for_training=np.load('train_features.npy')
client_features_with_id=np.load('client_dict_no_kp.npy',allow_pickle=True).item()
```

```python
images=[]
count=[]
for object in sift_features_with_id.keys():
    images.append(object)
    count.append(len(sift_features_with_id[object]))
print('Total Database Images : ',len(images))
print('Object with 2 images :', images[np.where(np.asarray(count)==2)[0][0]],'and',images[np.where(np.asarray(count)==2)[0][1]])
print('Object with 4 images :', images[np.where(np.asarray(count)==4)[0][0]])
print('All other objects have 3 images')
print('Total SIFT features used for training :',sift_features_for_training.shape[0])
print('Avg. feature per object :',sift_features_for_training.shape[0]/len(images))
print('Avg. feature per image :',sift_features_for_training.shape[0]/sum(count))
```

```python
images=[]
count=[]
totalfeat=0
for object in client_features_with_id.keys():
    images.append(object)
    count.append(len(client_features_with_id[object]))
    totalfeat+=client_features_with_id[object][0][1].shape[0]

print('Total Query Images : 50 , one image per object')
print('Total SIFT features : ', totalfeat)
print('Average SIFT features per object :',totalfeat/50)
```

```python
b_d_pair=[(4,3),(4,5),(5,7)]
for i in range(len(b_d_pair)):
    vt=VocabularyTree(b_d_pair[i][0],b_d_pair[i][1],sift_features_for_training)
    vt.build_tree()
    vt.generate_database_index(sift_features_with_id)
    tot1=0
    tot5=0
    for f in client_features_with_id:
        scores=vt.evaluate_querry(client_features_with_id[f][0][1])
        if f in list(scores.keys())[0:5]:
            tot5+=1
        if f == list(scores.keys())[0]:
            tot1+=1
    print('Branch : ',b_d_pair[i][0],' and Depth : ',b_d_pair[i][1])
    print('average top-1:',tot1/50)
    print('average top-5:',tot5/50)
```
